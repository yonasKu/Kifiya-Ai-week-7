```python
import torch
import cv2
import os
from google.colab import drive

# Mount Google Drive to access your folder
drive.mount('/content/drive')

# Load pre-trained YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s')

# Path to the directory containing the images in Google Drive
image_folder = '/content/drive/MyDrive/Kifiya/week 7/photos'  # Update this with your folder path in Google Drive

# Loop through the images in the folder and run object detection
results_list = []
for img_name in os.listdir(image_folder):
    img_path = os.path.join(image_folder, img_name)

    # Read the image using OpenCV
    img = cv2.imread(img_path)

    if img is None:
        print(f"Failed to load {img_path}")
        continue

    # Run object detection
    results = model(img)
    results_list.append((img_name, results))  # Store the results for saving later

    # Optionally, you can show the results in real-time (remove the comment below if you want to see the images)
    results.show()
```

```python
import os
import pandas as pd

# Folder to save the detection results (both images and CSV files)
main_save_folder = '/content/drive/MyDrive/Kifiya/week 7/detection_results'
result_folder = os.path.join(main_save_folder, 'result')  # Create a folder called 'result'

# Make sure the result folder exists
os.makedirs(result_folder, exist_ok=True)

# Process results and save them
for img_name, results in results_list:  # Assuming results_list is already defined

    # Define the target image path for clarity
    target_img_path = os.path.join(result_folder, f'{img_name}_detected.jpg')

    # Save detected image with bounding boxes directly in the result folder
    results_img_path = results.save(save_dir=result_folder)  # Ensure this doesn't create subfolders

    # Extract detections into a pandas DataFrame
    detections = results.pandas().xyxy[0]  # DataFrame of detection results
    print(f"Detections for {img_name}:")

    # Check and print the relevant detection information
    for index, row in detections.iterrows():
        class_name = row['name']  # Class label
        confidence = row['confidence']  # Confidence score
        xmin = row['xmin']  # Minimum x-coordinate
        ymin = row['ymin']  # Minimum y-coordinate
        xmax = row['xmax']  # Maximum x-coordinate
        ymax = row['ymax']  # Maximum y-coordinate

        # Print the extracted information
        print(f"Class: {class_name}, Confidence: {confidence:.2f}, Bounding Box: [{xmin}, {ymin}, {xmax}, {ymax}]")

    # Save detections to CSV file in the 'result' folder
    csv_path = os.path.join(result_folder, f'{img_name}_detections.csv')
    detections[['name', 'confidence', 'xmin', 'ymin', 'xmax', 'ymax']].to_csv(csv_path, index=False)

print(f"Processed images and CSV files saved in: {result_folder}")
```
